import numpy(arrays), matplotlib.pylot(visualization), and pandas(matrices)

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

getting dataset from my drive, like last hw divide dataset to 2 vectors for independent vs dependent

```python
dataset = pd.read_csv('/content/drive/My Drive/Position/Position_Salaries_poly.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

this trains the dataset we loaded and using linear regression instead of x and y train/test

```python
from  sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

The linear regression is a bit off and doesn't fit the trend

```python
plt.scatter(X, y, color = 'red')
X_lin_pred = lin_reg.predict(X)
plt.plot(X, X_lin_pred, color = 'black')
plt.title('Truth of Bluff (Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

using polynomial regression in order to train our dataset. using preprocessing function poly_reg.fit_transform in order to convert to a specific power.

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

here I used the polynomial regression

```python
X_poly = poly_reg.fit_transform(X)
X_lin_pred = lin_reg_2.predict(X_poly)
plt.scatter(X, y, color = 'red')
plt.plot(X, X_lin_pred, color = 'black')
plt.title('Truth of Bluff (Polynomial Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```
