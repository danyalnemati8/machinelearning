```python
df = pd.read_csv('/content/drive/My Drive/nba/nba_games.csv', index_col=0)
```

```python
df = df.sort_values("date")
```

```python
df = df.reset_index(drop=True)
```

```python
df
```

```python
del df["mp.1"]
del df["mp_opp.1"]
del df["index_opp"]
df
```

```python
def add_target(team):
  team["target"] = team["won"].shift(-1)
  return team

df = df.groupby("team", group_keys= False).apply(add_target)
df[df["team"] == "LAL"]
```

```python
df["target"][pd.isnull(df["target"])] = 2
```

```python
df["target"] = df["target"].astype(int,errors="ignore")
df
```

```python
nulls = pd.isnull(df)
```

```python
nulls = nulls.sum()
```

```python
nulls = nulls[nulls > 0]
nulls
```

remvoing extra columns that I do not need/ columns with null

```python
valid_columns = df.columns[~df.columns.isin(nulls.index)]
valid_columns
```

making a copy in order to not get errors in the long run since it will be chopped up

```python
df = df[valid_columns].copy()
df
```

TimeSeriesSplit makes sure that all the data from an older date is used to predict not the new games.

```python
from sklearn.model_selection import TimeSeriesSplit
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import RidgeClassifier

rr = RidgeClassifier(alpha=1)
split= TimeSeriesSplit(n_splits = 3)
sfs = SequentialFeatureSelector(rr, n_features_to_select=30, direction="forward", cv=split)
```

```python
removed_columns = ["season", "date", "won", "target", "team", "team_opp"]
```

```python
selected_columns = df.columns[~df.columns.isin(removed_columns)]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df[selected_columns] = scaler.fit_transform(df[selected_columns])
```

```python
df
```

```python
sfs.fit(df[selected_columns], df["target"])
```
