import numpy(arrays), matplotlib.pylot(visualization), and pandas(matrices)

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

getting dataset from my drive, like last hw divide dataset to 2 vectors for independent vs dependent

```python
dataset = pd.read_csv('/content/drive/My Drive/Linear_Reg/Salary_Data_simple lin.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

training set and testing set: X_train/test, y_train/test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/6, random_state = 3)
```

imported linearRegresssion and initilized regressor, using X and y train in the fit function

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

using predit function for the X_test from LinearRegression

```python
y_pred = regressor.predict(X_test)
```

plotting X_train and y_train and used predict to plot X_train

```python
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'black')
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years')
plt.ylabel('Salary')
plt.show()
```

plotting X_test and y_test and using predict o plot X_train, so pretty much the same as the graph above but displaying X_test and y_test instead of the trains

```python
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'black')
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years')
plt.ylabel('Salary')
plt.show()
```
