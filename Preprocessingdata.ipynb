imorted numpy,matplotlib, and pandas

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

loaded the given dataset  abd made it into two vectors independent vs dependent

```python
dataset = pd.read_csv('/content/drive/My Drive/hw2/Data.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
print(X)
```

```python
print(y)
```

imported sklearn for the missing values

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(X[:, 1:3])
X[:, 1:3] = imputer.transform(X[:, 1:3])
```

mean was repalced with nan

```python
print(X)
```

This code block utilizes OneHotEncoder to encode independent variables beginning at index 0.

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
print(X)
```

code ciohgers the dependentvariable using sklearn

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
print(y)
```

basically training and testing (x,y)_train, (x,y)_test separatley

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```python
print(X_test)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train[:, 3:] = sc.fit_transform(X_train[:, 3:])
X_test[:, 3:] = sc.transform(X_test[:, 3:])
```
